```python
import json
import pandas as pd
import random

# load your file
with open(os.path.join(path, 'intents.json'), "r", encoding="utf-8") as f:
    data = json.load(f)

pairs = []

# intents list
intents = data["intents"]

for intent in intents:
    patterns = intent.get("patterns", [])
    responses = intent.get("responses", [])

    # one-to-one mapping (cycle responses if needed)
    for i, pattern in enumerate(patterns):

        response = responses[i % len(responses)]  # safe mapping

        pairs.append({
            "user": pattern,
            "assistant": response,
            "tag": intent.get("tag")
        })

# convert to DataFrame
df = pd.DataFrame(pairs)

print(df.head())

# save
# df.to_csv("chat_dataset.csv", index=False)
```

```python
df
```

```python
def add_tag_user(text=""):
    return f"User: {text}\n"


def add_tag_assistant(text=""):
    return f"Assistant: {text}<|endoftext|>\n"


df['user'] = df['user'].apply(add_tag_user)
df['assistant'] = df['assistant'].apply(add_tag_assistant)
```

```python
df
```

```python
df["text"] = df["user"] + df["assistant"]
```
