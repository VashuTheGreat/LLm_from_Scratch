```
import torch

import torch.nn as nn
```

```
vocab_size=50257
output_dim=256


token_embedding_layer=nn.Embedding(vocab_size,output_dim)
```

```
batch_size=8
context_size=4
token_embedding=token_embedding_layer(torch.randint(0, vocab_size, (batch_size,context_size),dtype=torch.long))

token_embedding.shape
```

```
positional_embeder=nn.Embedding(context_size,output_dim)
```

```
posi_encodings=positional_embeder(torch.arange(context_size))

posi_encodings.shape
```

```
final_embeddings=token_embedding+posi_encodings

final_embeddings.shape
```
