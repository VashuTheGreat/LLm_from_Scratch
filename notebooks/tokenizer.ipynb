```python
from typing import List
import re

class SentenceTokenizerV1:
  def __init__(self,vocab:dict):
    self.int_to_str=vocab
    self.str_to_int={v:k for k,v in vocab.items()}


  def encode(self,text:str):
    text=re.split(r'([,.:;?_!"()\']|--|\s)',text)
    text=[r for r in text if r.strip()]
    # Correcting the syntax for handling unknown tokens
    text=[t if t in self.str_to_int else '<|unk|>' for t in text]

    tokens=[self.str_to_int[t] for t in text]

    return tokens

  def decode(self,tokens:List[int]):

    text=" ".join([self.int_to_str[t] for t in tokens])

    text = re.sub(r'\s+([,.?!"()\[\]{}])', r'\1', text)
    text = re.sub(r'\s+(--)\s*', r'\1', text)
    text = re.sub(r'\s+(\'s|\'t|\'re|\'ve|\'ll|\'d)', r'\1', text)

    return text
```

# Coding Input_Output Pairs

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class GPTDatasetV1(Dataset):
  def __init__(self,txt,tokenizer,max_len,stride):

    self.input_ids=[]
    self.output_ids=[]


    token_ids=tokenizer.encode(txt,allowed_special={"<|endoftext|>"})
    for i in range(0,len(token_ids)-max_len,stride):
      self.input_ids.append(token_ids[i:i+max_len])
      self.output_ids.append(token_ids[i+1:i+max_len+1])


  def __len__(self):
    return len(self.input_ids)

  def __getitem__(self,idx):
    return {"input_ids":self.input_ids[idx],"output_ids":self.output_ids[idx]}
```

```python
GPTDatasetV1(txt=text,tokenizer=tokenizer,max_len=4,stride=1)[0]
```

```python
def create_GPTDatasetV1_loader(txt,max_len,stride,batch_size,shuffle=True,drop_last=True,num_workers=0):
  tokenizer=tiktoken.get_encoding("gpt2")
  data_loader=DataLoader(GPTDatasetV1(txt=txt,tokenizer=tokenizer,max_len=max_len,stride=stride),batch_size=batch_size,shuffle=shuffle,drop_last=drop_last,num_workers=num_workers)
  return data_loader
```

```python
dataset_loader=create_GPTDatasetV1_loader(txt=text,max_len=4,stride=1,batch_size=2)
```
